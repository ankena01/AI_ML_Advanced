##### Data Ingestion

```python
from sklearn import datasets
```

```python
diabetes = datasets.load_diabetes()
```

```python
dir(diabetes)
```

```python
print(diabetes['DESCR'])
```

```python
diabetes['data'][0]
```

```python
diabetes['feature_names']
```

```python
diabetes['target']
```

```python
import pandas as pd
```

```python
X = pd.DataFrame(data = diabetes['data'], columns = diabetes['feature_names'])
```

```python
X
```

```python
y = pd.DataFrame(diabetes['target'], columns=['Blood Sugar Level'])
```

```python
y
```

```python
df = pd.concat([X,y] , axis = 1)
```

```python
df
```

```python
df['Target'] = df['Blood Sugar Level'].apply(lambda x: 'Normal' if x <100 else 'Possible Diabetic')
```

```python
df
```

##### Convert from categorical to neumweric data using Label Encodera

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le_Target = le.fit_transform(df['Target'])
```

```python
df['Target Encoded'] = le_Target
```

```python
df
```

```python
final_df = df.drop(['Blood Sugar Level','Target'] , axis =1 )
```

```python
X = final_df.drop('Target Encoded', axis = 1)
y = final_df['Target Encoded']
```

```python
X
```

```python
y
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```
