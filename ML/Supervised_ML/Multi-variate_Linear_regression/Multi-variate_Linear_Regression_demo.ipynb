### Data Pre-processing to handle null values

```python
df['bedrooms'].fillna(df['bedrooms'].median(), inplace= True)
```

```python
### Method 2

df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
```

### Create a linear regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()   ##### Hyper parameter tunning to be applied at this stage if needed
```

### Model Training on historical data

```python
df.columns
```

```python
X = df[['area', 'bedrooms', 'age']]
X
```

```python
y = df['price']
y
```

```python
model.fit(X , y )
```

### Model Evaluation

```python
model.predict([[4100,6,8]])
```

### Slope and Y-Intercept

```python
m = model.coef_
m
```

```python
b = model.intercept_
b
```
