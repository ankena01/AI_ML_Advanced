#### Data Ingestion

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Admin\Documents\DataAnalysis\1py-master\py-master\ML\5_one_hot_encoding\homeprices.csv")
df
```

#### Pandas Get Dummies method - To handle categorical data

```python
dummies = pd.get_dummies(df['town'])
dummies
```

### Adding the Neumerical data into orignal dataframe

```python
merged_df = pd.concat([df,dummies], axis = 1)
merged_df
```

### Remove the town column 

```python
df_2 = merged_df.drop('town', axis = 1)
```

```python
df_2
```

```python
### define independent feature

X = df_2.drop('price' , axis = 1)
X
```

```python
### define target variable

y = df_2['price']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
y_pred
```

```python
df_2['Predicted Price'] = y_pred
```

```python
df_2
```

```python
model.score(X,y)
```
